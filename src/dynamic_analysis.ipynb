```python
import numpy as np
import matplotlib.pyplot as plt
from thermalspin.data_analysis  import *
```

```python
# Insert here the name of the simulation set
simname = "asdasd"
```

```python
final_state, t, J, D, Hz, T, e, m = load_results(simname)
```

```python
fig = plt.figure(figsize=(8,5))
ax = plt.subplot(111)
ax.plot(t,e)
plt.xlabel("Step")
plt.ylabel("Energy")
plt.grid()
plt.show()
plt.savefig("plots/hysteresis_energy.svg")
```

```python
fig = plt.figure(figsize=(10,6))
ax = plt.subplot(111)
ax.plot(t, m[:,0], label = r"$M_x$")
ax.plot(t, m[:,1], label = r"$M_y$")
ax.plot(t, m[:,2], label = r"$M_z$")
plt.grid()
plt.legend()
plt.ylabel("Magnetization")
plt.xlabel("Steps")
plt.title("Magnetization")
plt.show()
plt.savefig("plots/hysteresis_magnetization.svg")
```

```python
fig = plt.figure(figsize=(10,6))
ax = plt.subplot(111)
ax.plot(t, m[:,2], label = r"$M_z$")
ax.plot(t, Hz, label = r"$H_z$")
ax.set_xlim([-0.01e8, 1.5e8])
plt.grid()
plt.legend()
plt.xlabel("Steps")
plt.title("Magnetization and external field")
plt.show()
plt.savefig("plots/hysteresis_MH.svg")
```

```python
fig = plt.figure(figsize=(10,6))
ax = plt.subplot(111)
ax.plot(Hz, m[:,2])
ax.grid()
ax.set_xlim([-0.8, 0.8])
ax.set_ylim([-1, 1])
plt.xlabel(r"$H_z$")
plt.ylabel(r"$M_z$")
plt.title("Histersys cycle")
plt.show()
plt.savefig("plots/hysteresis_cycle.svg")
```
